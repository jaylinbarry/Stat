```python
import numpy as np
import matplotlib.pyplot as plt
#importing numpy and matplotlib to use in oru codes
```

```python
x = np.linspace(0,10,50)
#defining x as the data from above 
```

```python
noise = 10 * np.random.randn(50,)
#defining noise as the data from above multiplied by 10
```

```python
y = 3 * x + 7 + noise
#defining y as 3 times the random value of x plus 7 plus the random number times 10 from above
```

```python
plt.plot(x,y,'.');
#graphing the data above into a plot graph
```

```python
m_guess = 1
b_guess = 1

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
#graphing the data from cx and y as above and defining m guess and b guess as 1 to be used in the graph
```

```python
m_guess = 5
b_guess = 15

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
#graphing the same data from above but with m guess as 5 and b guess as 15, instead of 1 
```

```python
m_guess = 3
b_guess = 7

y_guess = m_guess * x + b_guess

plt.plot(x,y_guess,'r.');
plt.plot(x,y,'g.');
#graphing the data again but with m guess as 3 and b guess as 7 
```

```python
loss = np.sum((y_guess - y)**2)
#defining loss as the sum of y guess minus y times 2
```

```python
loss
#having the computer show us the value of loss as we just defined
```

```python
for m_guess in np.linspace(-10,10,10):
    for b_guess in np.linspace(-10,10,10):
        
        y_guess = m_guess * x + b_guess

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,m_guess,b_guess)
        #graphing all of the data from above into seperate graphs 
```

```python
best_loss = 100000
best_m = 0
best_b = 0 

for m_guess in np.linspace(-10,10,20):
    for b_guess in np.linspace(-10,10,20):
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 

            plt.plot(x,y_guess,'r.');
            plt.plot(x,y,'g.');
            plt.show()

            print(loss,m_guess,b_guess)
            #graphing all of the data into seperate graphs using loss m and b in each
```
